```python
# Always run these two lines first
from symbulate import *
%matplotlib inline
```

```python
n = 4  # number of babies
labels = list(range(n))  # This will label the babies starting with 0:  0, 1, 2, 3
P = BoxModel(labels, size = n, replace = False)

shuffles = P.sim(10000)

shuffles # print the output
```

```python
# tabulate results results of many shuffles - counts
shuffles.tabulate()
```

```python
# define a function that takes a shuffle as input and returns whether there are no matches
def is_match(x):
    for i in range(n):
        if x[i] == labels[i]:
            return 'At least one match'
    return 'No match'

# now apply that function to each of the shuffles in the simulation
shuffles.apply(is_match)
```

```python
shuffles.apply(is_match).tabulate() # Apply no_match to all shuffles and tabulate
```

```python
shuffles.apply(is_match).tabulate(normalize = True) # Apply no_match to all shuffles and tabulate
```
